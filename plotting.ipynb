```python
%matplotlib inline

import time

import matplotlib.pyplot as plt
import matplotlib.tri as tri
import netCDF4
import numpy as np
```

```python
class Field(object):
    def __init__(self, node_values, node_coordinates, face_node_connectivity):
        self.node_values = np.array(node_values)
        self.node_coordinates = np.array(node_coordinates)
        self.face_node_connectivity = np.array(face_node_connectivity)

    def __getitem__(self, keys):        
        if isinstance(keys, int):
            keys = [keys]
        keys = np.array(keys)
        
        # Restrict the face-node connectivity to the selected faces,
        # but still indexing into the full set of nodes.
        sub_face_node_connectivity = self.face_node_connectivity[[keys]]
        
        # Which nodes do we still need?
        node_indices = np.unique(sub_face_node_connectivity)
        
        # Restrict the node values & coordinates to just the needed nodes.
        sub_node_values = self.node_values[node_indices]
        sub_node_coordinates = self.node_coordinates[node_indices]

        # Figure out the mapping from old node indices to new node indices.
        node_map = np.zeros(len(self.node_coordinates), int)
        node_map[node_indices] = np.arange(len(node_indices), dtype=int)

        # Update the face-node connectivity to use the new node indices.
        sub_face_node_connectivity = node_map[sub_face_node_connectivity]

        return Field(node_values=sub_node_values,
                     node_coordinates=sub_node_coordinates,
                     face_node_connectivity=sub_face_node_connectivity)

    def triangulate(self):
        triangulation = tri.Triangulation(self.node_coordinates[:, 0],
                                          self.node_coordinates[:, 1],
                                          self.face_node_connectivity)
        return triangulation

    def contourf(self, triangulation):
        plt.tricontourf(triangulation, self.node_values, cmap='Blues')

    def __repr__(self):
        return ('Field(node_values={!r},\n'
                '    node_coordinates={!r},\n'
                '    face_node_connectivity={!r})'.format(
                    self.node_values, self.node_coordinates, 
                    self.face_node_connectivity))
```

```python
dataset = netCDF4.Dataset('SELFE.nc')
phenom = dataset.variables['depth'][:]
node_x = dataset.variables['x'][:]
node_y = dataset.variables['y'][:]
face_node_connectivity = dataset.variables['ele'][:].T - 1  # Convert from 1-based to 0-based indexing.
```

```python
coords = np.vstack([node_x, node_y]).T
field = Field(node_values=phenom,
              node_coordinates=coords,
              face_node_connectivity=face_node_connectivity)
```

```python
triangulation = field.triangulate()
field.contourf(triangulation)
```

```python
sub_field = field[np.arange(100000)]
triangulation = sub_field.triangulate()
sub_field.contourf(triangulation)
```

```python
n_tris = [100, 1000, 10000, 30000, 100000, 500000, 800000, 1000000, 1166747]
draw_timings = []
plt.gca()
plt.gcf().canvas.draw()
for n in n_tris:
    sub_field = field[np.arange(n)]
    triangulation = sub_field.triangulate()
    t0 = time.time()
    sub_field.contourf(triangulation)
    plt.gcf().canvas.draw()
    delta = time.time() - t0
    plt.clf()
    draw_timings.append(delta)

plt.plot(n_tris, draw_timings, marker='o')
plt.xlabel('# of triangles')
_ = plt.ylabel('Time to draw / s')
```
